```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
s = pd.read_excel ('/content/Resultados GSMW.xlsx')
map = pd.DataFrame(s, columns= ['Windows Length','Windows Start','Q2'])
print (map)

#x=x.values.tolist()
```

```python
# prompt: realizar un mapa de color de windows length y windows start coloreados por Q2 usando cmap orange. no uses map.pivot, con espaciado de 10 para la grilla
plt.rcParams.update({'font.size': 17})


# Assuming 'map' DataFrame is already created as in your original code

# Create the heatmap
plt.figure(figsize=(8, 6))
plt.imshow(map.pivot(index='Windows Length', columns='Windows Start', values='Q2'), cmap='Oranges', aspect='auto')
plt.colorbar(label='Q2')
plt.xlabel('Windows Start')
plt.ylabel('Windows Length')


# Customize the grid
plt.xticks(np.arange(0, len(map['Windows Start'].unique()), 10), rotation=90)
plt.yticks(np.arange(0, len(map['Windows Length'].unique()), 20))
plt.grid(True, which='both', color='grey', linestyle='--')  # Add a grid

plt.savefig('Fig 15.png', dpi=1200, bbox_inches = 'tight')
plt.show()

```
